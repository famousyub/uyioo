```python
import imdb

def get_movie_id(movie_name):
    ia = imdb.IMDb()
    movies = ia.search_movie(movie_name)
    if movies:
        movie = movies[0]
        return movie.movieID
    else:
        return None



movie_name = "interstellar"
movie_id = get_movie_id(movie_name)
print(movie_id)

res=f"tt{movie_id}"

res
```
